```python
import numpy
import math
from astropy.io import fits
from astropy.table import Table
from matplotlib.colors import LogNorm
from astropy.modeling import models, fitting

#set up matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DAP_Pipe3D = fits.open("DAP_Pipe3D.fits", memmap=True)
DAP_Pipe3D_table = Table(DAP_Pipe3D[1].data)
```

```python

SFR_DAPinitial = numpy.array(DAP_Pipe3D_table['SFR_TOT'])  # Stores all data of SFR DAP (h-2 Msun/yr)
SFR_Pipe3Dinitial = numpy.array(DAP_Pipe3D_table['log_SFR_Ha']) # Stores all data of SFR Pipe3D  log(Msun/yr)

h = 0.71

SFR_DAP = SFR_DAPinitial * h**(-2)         # SFR of DAP in Msun/yr
SFR_Pipe3D = 10 ** SFR_Pipe3Dinitial    # SFR of Pipe3D in Msun/yr

SFR_DAP_log = numpy.log10(SFR_DAPinitial * h**(-2))      # SFR of DAP in log(Msun/yr)
SFR_Pipe3D_log = SFR_Pipe3Dinitial          # SFR of Pipe3D in log(Msun/yr)

print("Max SFR in DAP", '\t', '\t', numpy.max(SFR_DAP))
print("Max SFR in Pipe3D", '\t', numpy.max(SFR_Pipe3D))

print(numpy.min(SFR_DAP))
print(numpy.min(SFR_Pipe3D))

print(SFR_DAP)
print(SFR_Pipe3D)
```

```python
# With Limits (Actual value)

plt.xlabel("SFR_DAP (Msun/yr)")
plt.ylabel("SFR_Pipe3D (Msun/yr)")
plt.title("Comparison of SFRs of from DAP and Pipe3D.")


plt.plot(SFR_DAP, SFR_Pipe3D, "b,") 
plt.plot(SFR_DAP, SFR_DAP, "k-")

# plt.xlim(0, 50)
# plt.ylim(0, 50)

plt.xlim(-0.01, 1)
plt.ylim(-0.01, 1)

plt.show()
```

```python
# With Limits (Log value)

plt.xlabel("SFR_DAP/ log(Msun/yr)")
plt.ylabel("SFR_Pipe3D/ log(Msun/yr)")
plt.title("Comparison of SFRs of from DAP and Pipe3D.")


plt.plot(SFR_DAP_log, SFR_Pipe3D_log, "b,") 
plt.plot(SFR_DAP_log, SFR_DAP_log, "k-")

plt.xlim(-3, 1)
plt.ylim(-5, 2)

plt.show()
```
