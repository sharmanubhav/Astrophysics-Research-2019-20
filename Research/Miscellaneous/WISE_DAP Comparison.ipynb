```python
import numpy
import math
from astropy.io import fits
from astropy.table import Table
from matplotlib.colors import LogNorm
from astropy.modeling import models, fitting

#set up matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
WISE_DAP = fits.open("WISE_DAP.fits", memmap=True)
WISE_DAP_table = Table(WISE_DAP[1].data)
```

```python
W1 = numpy.array(WISE_DAP_table['W1mag'])  # Stores all data of W1
W2 = numpy.array(WISE_DAP_table['W2mag'])  # Stores all data of W2
W3 = numpy.array(WISE_DAP_table['W3mag'])  # Stores all data of W3
W4 = numpy.array(WISE_DAP_table['W4mag'])  # Stores all data of W4
SFR = numpy.array(WISE_DAP_table['SFR_TOT']) # Stores all data of SFR_TOT

W1diffW2 = W1 - W2
W2diffW3 = W2 - W3

# SFRfactor = (0.71) ** (-2) * 1.989 * 10 ** 30   # h^-2 * Mass of Sun 

    
lowDiff = numpy.where(W2diffW3 < 2)



diff2line = numpy.full(len(W1), 2)
```

```python
plt.xlabel("W2-W3/ mag")
plt.ylabel("W1-W2/ mag")
plt.title("WISE mid-infrared comparison for entire sample (DAP)")


plt.plot(W2diffW3, W1diffW2, "g.") 
plt.plot(diff2line, W1diffW2, "b-.")

plt.xlim(0, 4)
plt.ylim(-0.2, 0.3)

plt.show()
```

```python
plt.xlabel("W2-W3/ mag")
plt.ylabel("SFR_TOT/ (h-2 Msun/yr)")
plt.title("WISE vs SFR_TOT for entire sample (DAP)")


plt.plot(W2diffW3, SFR, "g.") 

plt.show()
```

```python
print("Max SFR of W2diffW3 < 2 : ", "\t", numpy.max(SFR[lowDiff]))
print("Max SFR of all : ", "\t", "\t", numpy.max(SFR)) 
```
